```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('sales_data.csv')
print(df)
```

```python
# Set Seaborn theme
sns.set(style='whitegrid')
```

```python
# Line plot: Sales over months
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Month', y='Sales', marker='o')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
# Bar plot: Profit by month
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Month', y='Profit', hue='Category')
plt.title('Monthly Profit by Category')
plt.show()
```

```python
# Box plot: Profit distribution by Category
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Category', y='Profit')
plt.title('Profit Distribution by Category')
plt.show()
```

```python
# Scatter plot: Sales vs Profit
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', s=100)
plt.title('Sales vs Profit')
plt.show()
```

```python
# Heatmap: Correlation matrix
plt.figure(figsize=(6, 4))
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```
